```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv(r"../zasoby/high_school_sat_gpa.csv",
                   sep=' ', usecols=['math_SAT','verb_SAT','high_GPA'])
 
data.head()
data.dtypes
```

```python
plt.scatter(data.math_SAT, data.high_GPA)
```

```python
regression = LinearRegression()
regression.fit(data.math_SAT.to_numpy().reshape(-1, 1), data.high_GPA.to_numpy().reshape(-1, 1))
```

```python
plt.scatter(data.math_SAT, data.high_GPA, c='blue')
plt.plot([data.math_SAT.min(), data.math_SAT.max()], [regression.predict(data.math_SAT.min().reshape(-1, 1))[0],
                                                      regression.predict(data.math_SAT.max().reshape(-1, 1))[0]],
        c='red')
```

```python
regression2 = LinearRegression()
regression2.fit(data.verb_SAT.to_numpy().reshape(-1,1), data.high_GPA.to_numpy().reshape(-1, 1))
```

```python
plt.scatter(data.verb_SAT, data.high_GPA, c='blue')
plt.plot([data.verb_SAT.min(), data.verb_SAT.max()], [regression2.predict(data.verb_SAT.min().reshape(-1, 1))[0],
                                                      regression2.predict(data.verb_SAT.max().reshape(-1, 1))[0]],
        c='red')
```

```python
regression3 = LinearRegression()
regression3.fit(data.loc[:, ['math_SAT', 'verb_SAT']].to_numpy(), data.high_GPA)
```

```python
plt.scatter(data.verb_SAT, data.high_GPA, c=['blue'])
plt.scatter(data.math_SAT, data.high_GPA, c=['green'])
plt.plot([data.verb_SAT.min(), data.math_SAT.max()], [regression3.predict(np.array([data.math_SAT.min(), data.verb_SAT.min()]).reshape(1, 2)),
                                                      regression3.predict(np.array([data.math_SAT.max(), data.verb_SAT.max()]).reshape(1, 2))],
        c='red')
```
